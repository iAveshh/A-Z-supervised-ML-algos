### Linear Regression with Least squares error.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

x, y  = make_regression(n_samples = 100, n_features=1,
                            n_informative=1, bias = 150.0,
                            noise = 30, random_state=0)
plt.scatter(x, y ,marker = 'o', s = 50)
```

```python
# now, lets create a linear Regression line over our data points

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(x,y, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
model = lin_reg.fit(X_train,y_train)
accu_r1 = model.score(X_train,y_train)
model = lin_reg.score(X_test,y_test)

print('linear model coeff (w): {}'
     .format(lin_reg.coef_))
print('linear model intercept (b): {:.3f}'
     .format(lin_reg.intercept_))
print('R-squared score (training): {:.3f}'
     .format(lin_reg.score(X_train, y_train)))
print('R-squared score (test): {:.3f}'
     .format(lin_reg.score(X_test, y_test)))
```

```python
plt.scatter(x, y ,marker = 'o', s = 50,alpha = 0.7)
plt.plot(x,lin_reg.coef_ * x+lin_reg.intercept_ , 'r-', alpha = 0.7)
plt.xlabel("Feature Values")
plt.ylabel("Target values")
plt.title("Least Square Linear Regression")
```
